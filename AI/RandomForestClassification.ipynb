```python
data = pd.read_csv('Airline.csv')

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['satisfaction'] = encoder.fit_transform(data['satisfaction'])
encoder.classes_


data.drop(columns=['Departure Delay in Minutes', 'Food and drink'], axis =1 , inplace=True)
data.describe()

data.drop('Departure/Arrival time convenient', axis = 1, inplace = True)
data.drop('Gate location', axis = 1, inplace = True)
data.drop('Flight Distance', axis = 1, inplace = True)
data[data.columns[data.dtypes=='O']] = data[data.columns[data.dtypes=='O']].astype(str).apply(LabelEncoder().fit_transform)
data.head()


data.dropna(inplace=True)

from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(data, test_size=0.3, random_state=42)

train_x = train_df.drop('satisfaction', axis=1)
train_y = train_df['satisfaction']

test_x = test_df.drop('satisfaction', axis=1)
test_y = test_df['satisfaction']


model_RFC = RandomForestClassifier().fit(train_x, train_y)
pred_RFC = model_RFC.predict(test_x)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, mean_squared_error
import numpy as np

# 성능 지표 계산
accuracy = accuracy_score(test_y, pred_RFC)
precision = precision_score(test_y, pred_RFC, average='weighted')
recall = recall_score(test_y, pred_RFC, average='weighted')
f1 = f1_score(test_y, pred_RFC, average='weighted')
confusion = confusion_matrix(test_y, pred_RFC)

# 성능 지표 출력
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(f'Confusion Matrix: \n{confusion}')
# MSE와 RMSE 계산
mse = mean_squared_error(test_y, pred_RFC)
rmse = np.sqrt(mse)

# 성능 지표 출력
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(f'Confusion Matrix: \n{confusion}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')




#테스트할 데이터
pred_src_data = pd.read_csv('Airline_test.csv')
pred_src_data.drop(columns=['Departure Delay in Minutes', 'Food and drink'], axis =1 , inplace=True)
pred_src_data.describe()

pred_src_data.drop('Departure/Arrival time convenient', axis = 1, inplace = True)
pred_src_data.drop('Gate location', axis = 1, inplace = True)
pred_src_data.drop('Flight Distance', axis = 1, inplace = True)
#범주형 데이터 처리
pred_src_data[pred_src_data.columns[pred_src_data.dtypes=='O']] = pred_src_data[pred_src_data.columns[pred_src_data.dtypes=='O']].astype(str).apply(LabelEncoder().fit_transform)
pred_src_data.head()
pred_src_data.dropna(inplace=True)


import pickle
# 모델 저장
with open('random_forest_model.pkl', 'wb') as file:
    pickle.dump(model_RFC, file)


# 모델 불러오기
with open('random_forest_model.pkl', 'rb') as file:
    model_RFC_loaded = pickle.load(file)

# 불러온 모델로 예측
pred_RFC_final = model_RFC_loaded.predict(pred_src_data)
y_pred = encoder.inverse_transform(pred_RFC_final)

pred_src_data['pred']=y_pred
pred_src_data.head()

# Save the results to a new CSV file
pred_src_data.to_csv('TEST_with_predictions.csv', index=False)


```
